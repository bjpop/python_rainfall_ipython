File objects keep track of their current position in the file, it is an offset from the start of the file, measured in bytes. Reads from (and writes to) the file happen at the current position, causing the position to be updated. Subsequent reads and writes happen at the new position.

When a file is opened the current position is set to the start of the file (offset 0).

You can ask for the current position using the `tell` method of the file object. You can set it to a new position using the `seek` method.

`tell` and `seek` are quite low-level and you don't normally need to use them directly.

## Collect all the entries for the month of January from the entire data set

We will accumulate the January entries in a list. Initially it is empty.

We use a `for` loop to iterate over the list of lines. Each line is a string.

For each line we split it into fields using the `split` method.

Grab the specific field for January on the current line, and append it to our collection of entries.

```python
# intialise a list to contain the entries for Jan, it is empty at the start
jan_entries = []
# process each line at a time
for line in lines:
    # split each line on the comma character
    fields = line.split(',')
    # select the value for January, it occurs at index 3 in the list of fields
    this_field = fields[3]
    # append this to our collection of Jan fields
    jan_entries.append(this_field)
jan_entries
```

## Compute the average rainfall for January over all years

Note that we have collected a list of entries for January. However, there are two issues we will have to deal with before we can compute the average:

1. The list contains strings, but we need a list of numbers (floating point numbers, actually)
2. Some of the fields are strings which do not represent numbers: 'Jan' and 'null'.

We can use the `float` method to turn a string into a floating point number. 

```python
float(3.14)
```

However, if we apply `float` to a string which does not represent a number then we will get a `ValueError` exception:

```python
float('Jan')
```

### Use exception handling to handle the `ValueError`

Fortunately we can gracefully deal with the problematic input by catching the `ValueError` with an exception handler:

```python
example_str = 'Jan'
try:
    result = float(example_str)
except ValueError:
    print('"{}" is not a valid string for float conversion'.format(example_str))
else:
    print('"{}" works fine: {}'.format(example_str, result))
```

```python
example_str = '3.142'
try:
    result = float(example_str)
except ValueError:
    print('"{}" is not a valid string for float conversion'.format(example_str))
else:
    print('"{}" works fine: {}'.format(example_str, result))
```

## Iterate over the rainfall for January and compute an average

```python
total = 0.0
count = 0
for entry in jan_entries:
    try:
        float_entry = float(entry)
    except ValueError:
        # we just skip this particular entry
        pass
    else:
        # In this case we didn't have an exception
        total += float_entry
        count += 1
jan_average = total / count
jan_average
```

### Print the average rainfall to 3 decimal places

```python
"The average rainfall for January was {:.3f}".format(jan_average)
```

## Write a helper function for computing averages

We are going to compute averages of lists of numbers many times over. 

The number one rule in programming is **Don't repeat yourself**. So we should define this common operation as a function.

If the input list is empty then we will raise a `ValueError` exception.

```python
def avg(numbers):
    count = len(numbers)
    if count > 0:
        return sum(numbers) / float(count)
    else:
        raise ValueError("avg: empty list as input")
```

```python
avg([4,5,6])
```

```python
avg([])
```

## Collect the January rainfall for every year, record a (year, measurement) pair

We have to be careful because some years do not have a meaningful rainfall measurement for January.

We'll also need to skip the first line because it is a header row.

For the sake of the example, we'll convert the year from a string to an integer. We'll assume that all the year values are valid.

```python
results = []
# skip the first line because it is a header row
data_lines = lines[1:]
for line in data_lines:
    fields = line.split(',')
    # convert the year into an integer, we assume it is valid
    year = int(fields[2])
    jan_measurement = fields[3]
    try:
        float_jan_measurement = float(jan_measurement)
    except ValueError:
        # we just skip this particular entry
        pass
    else:
        # record the year and the rainfall measurement
        results.append((year, float_jan_measurement))
print(results)
```

## Unpack the list of pairs into two lists, one containing years, the other containing measurements

The `pylab.plot` function requires our data for the X and Y axes to be in separate lists, rather than a list of pairs. Therefore we will need to unzip the results computed above.

One way to do this is to use a for loop:

```python
years = []
measurements = []
for (year, measurement) in results:
    years.append(year)
    measurements.append(measurement)
print(years)
print(measurements)
```

A tidier solution is to use list comprehensions:

```python
years = [year for (year, _) in results]
measurements = [measurement for (_, measurement) in results]
print(years)
print(measurements)
```

## Use the CSV library for reading the input data file

As we have shown above it is possible to read and parse a CSV file directly in Python without using a dedicated library.

However, this approach has some serious shortcomings:
    
1. We have to remember to treat the header line and the data lines differently.
2. It is difficult to support the field delimiter (such as the comma) when it appears inside quotation marks.

Fortunately Python comes with a library called `csv` which solves these problems, and many more.

```python
file.close()

import csv
file = open(filename)
```

The `csv` library provides various ways to read a CSV file. The simplest way is via the `reader` function. It parses the file and returns an iterator that yields one input row at a time, split on the delimiter character.

Each row is represented as a list of strings.

The default delimiter is the comma, but you can override that with an optional argument if necessary.

```python
reader = csv.reader(file)
for row in reader:
    print(row)
```

While the `reader` function is simple it does not provide any special functionality for handling the header row.

A much fancier and more powerful interface is provided by the `DictReader` function.

The `DictReader` assumes that the first row is a header row. It parses the file and returns an iterator that yields one dictionary for each input row at a time, split on the delimiter character. The keys of the dictionary correspond to the fields of the heading row. Each key in the dictionary is associated with its corresponding data value from the current row.

```python
file.close()
file = open(filename)
reader = csv.DictReader(file)
for row in reader:
    print(row)
```

## Plot the annual rainfall for each year

Again we will make the simplifying assumption that the years are valid integers, though we will check to make sure the annual rainfall can be parsed as a floating point number.

You should note that this code is much tider than our previous attempt at reading data from the CSV file. The main reason is that the `DictReader` does a lot of work on our behalf.

```python
file.close()
file = open(filename)
reader = csv.DictReader(file)
years = []
totals = []
for row in reader:    
    try:
        annual_float = float(row['Annual'])
    except ValueError:
        pass
    else:
        years.append(int(row['Year']))
        totals.append(annual_float)

annual_average = avg(totals)
min_year = years[0]
max_year = years[-1]
pylab.plot(years, totals)
pylab.plot((min_year, max_year), (annual_average, annual_average))
pylab.xlim(min_year, max_year)
pylab.xlabel('year')
pylab.ylabel('rainfall (mm)')
pylab.title('Annual rainfall against year')
```

## Generalise our code so that it can plot the rainfall for any of the columns in the data

We can write a function that generalises what we have done so far. We can give it a couple of parameters which make it usable in many circumstances:

1. A filename for the data set to process.
2. A column header for the attribute that we'd like to plot (it could be a given month, or the annual rainfall, for example).

```python
def plot_rainfall(filename, attribute):
    file = open(filename)
    reader = csv.DictReader(file)
    years = []
    values = []
    for row in reader:    
        try:
            value_float = float(row[attribute])
        except ValueError:
            pass
        else:
            years.append(int(row['Year']))
            values.append(value_float)
    # We are finished reading from the file, so let's close it
    file.close()
    values_average = avg(values)
    min_year = years[0]
    max_year = years[-1]
    pylab.plot(years, values)
    pylab.plot((min_year, max_year), (values_average, values_average))
    pylab.xlim(min_year, max_year)
    pylab.xlabel('year')
    pylab.ylabel('rainfall (mm)')
    pylab.title('{} rainfall against year'.format(attribute))
```

```python
plot_rainfall('rainfall.csv', 'Jan')
```

```python
plot_rainfall('rainfall.csv', 'Mar')
```

```python
plot_rainfall('rainfall.csv', 'Annual')
```
